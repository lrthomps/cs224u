```python
def load_dataset(*src_filenames, labels=None):
    data = []
    for filename in src_filenames:
        with open(filename) as f:
            for line in f:
                d = json.loads(line)
                if labels is None or d['gold_label'] in labels:
                    data.append(d)
    return data
```

```python
from transformers import BertModel, BertTokenizer
import vsm
```

```python
import torch
```

```python
bert_weights_name = 'bert-base-cased'
bert_tokenizer = BertTokenizer.from_pretrained(bert_weights_name)
bert_model = BertModel.from_pretrained(bert_weights_name)
```

```python
device = "cuda:0" if torch.cuda.is_available() else "cpu"
```

```python
bert_model = bert_model.to(device)
```

```python
def bert_phi(text):
    subtok_ids = hf_encode(text, bert_tokenizer)
    if len(subtok_ids) > 512:
      return None
    torch_array = vsm.hf_represent(subtok_ids.to(device), bert_model, layer=-1)
    return [torch_array[0, i, :].cpu().numpy() for i in range(torch_array.shape[1])]
```

```python
def build_Xy(df, phi):
    X = df['sentence'].apply(phi).tolist()
    y = df['label'].values
    return {'X': X, 'y': y}
```

```python
def hf_encode(text, tokenizer, add_special_tokens=False):
    """
    Get the indices for the tokens in `text` according to `tokenizer`.
    If no tokens can be obtained from `text`, then the tokenizer.unk_token`
    is used as the only token.

    Parameters
    ----------
    text: str

    tokenizer: Hugging Face tokenizer

    add_special_tokens : bool
        A Hugging Face parameter to the tokenizer.

    Returns
    -------
    torch.Tensor of shape `(1, m)`
        A batch of 1 example of `m` tokens`, where `m` is determined
        by `text` and the nature of `tokenizer`.

    """
    encoding = tokenizer.encode(
        text,
        add_special_tokens=add_special_tokens,
        return_tensors='pt', max_length=512, truncation=True)
    if encoding.shape[1] == 0:
        text = tokenizer.unk_token
        encoding = torch.tensor([[tokenizer.vocab[text]]])
    return encoding
```
